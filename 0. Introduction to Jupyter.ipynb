This was my "show off" script - a blend of python (for loops), data analytics (pandas), and logic - showing the power of what we can do when we add this stuff all together.

```python
# Let's do some fancy plotting - we'll loop through each restaurant, creating a different plot for one of two conditions
import matplotlib.pyplot as plt
import pandas as pd
df_summarized_orders = pd.read_csv('data/summarized_orders.csv')
# Some quick data formatting to get advanced date capabilities
df_summarized_orders["day"] =  pd.to_datetime(df_summarized_orders["day"])
# Loop through each restaurant
for r in df_summarized_orders["location"].unique().tolist():
    # Print the restaurant name
    print(r)
    
    # make the legend of the chart an empty list
    legend = []
    
    # Filter the orders down to this restaurant
    df = df_summarized_orders[df_summarized_orders["location"] == r]
    
    # separate weekdays and weekends
    weekday_rows = (df["day"].dt.weekday < 5)
    weekend_rows = (df["day"].dt.weekday >= 5)
    
    # Plot weekday and weekend histogram
    if weekday_rows.sum() > 0:
        df[weekday_rows]["number of orders"].plot(kind='hist', bins=30,
                                   density=True) 
        legend.append("Weekday")
    
    if weekend_rows.sum() > 0:
        df[weekend_rows]["number of orders"].plot(kind='hist', bins=30, 
                              density=True, alpha=0.5) 
        legend.append("Weekend")
    # Create a legend
    plt.legend(legend)
    
    # Show the plot at each stage of the loop
    # this looks for any active plt objects and displays them
    plt.show()
```
